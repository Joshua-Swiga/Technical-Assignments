```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# Set a random seed for reproducibility
np.random.seed(42)

# Generate sample size
n = 200

# Create independent variables with a linear trend
drug_encapsulation_efficiency = np.linspace(50, 100, n) + np.random.normal(0, 2, n)  # Positive trend
clearance_rate = np.linspace(1, 5, n) + np.random.normal(0, 0.1, n)  # Inverse trend
targeting_precision = np.linspace(50, 100, n) + np.random.normal(0, 2, n)  # Positive trend
volume_of_distribution = np.linspace(2, 10, n) + np.random.normal(0, 0.1, n)  # Positive trend
blood_flow_rate = np.linspace(50, 200, n) + np.random.normal(0, 2, n)  # Positive trend

# Create Drug Efficacy with directional dependence
drug_efficacy = (0.5 * drug_encapsulation_efficiency + 
                 0.2 * targeting_precision - 
                 0.5 * clearance_rate + 
                 0.1 * volume_of_distribution + 
                 0.05 * blood_flow_rate + 
                 np.random.normal(0, 5, n))  # Add some noise

# Combine into a DataFrame
data = pd.DataFrame({
    'Drug Encapsulation Efficiency (%)': drug_encapsulation_efficiency,
    'Clearance Rate (L/hr)': clearance_rate,
    'Targeting Precision (%)': targeting_precision,
    'Volume of Distribution (L)': volume_of_distribution,
    'Blood Flow Rate (mL/min)': blood_flow_rate,
    'Drug Efficacy (%)': drug_efficacy
})

# Save the dataset to a CSV file
data.to_csv('drug_study_data.csv', index=False)

# Visualization
plt.figure(figsize=(14, 10))

# Plot each variable against Drug Efficacy
plt.subplot(3, 2, 1)
sns.scatterplot(x='Drug Encapsulation Efficiency (%)', y='Drug Efficacy (%)', data=data)
plt.title('Drug Encapsulation Efficiency vs. Drug Efficacy')
plt.xlabel('Drug Encapsulation Efficiency (%)')
plt.ylabel('Drug Efficacy (%)')

plt.subplot(3, 2, 2)
sns.scatterplot(x='Clearance Rate (L/hr)', y='Drug Efficacy (%)', data=data)
plt.title('Clearance Rate vs. Drug Efficacy')
plt.xlabel('Clearance Rate (L/hr)')
plt.ylabel('Drug Efficacy (%)')

plt.subplot(3, 2, 3)
sns.scatterplot(x='Targeting Precision (%)', y='Drug Efficacy (%)', data=data)
plt.title('Targeting Precision vs. Drug Efficacy')
plt.xlabel('Targeting Precision (%)')
plt.ylabel('Drug Efficacy (%)')

plt.subplot(3, 2, 4)
sns.scatterplot(x='Volume of Distribution (L)', y='Drug Efficacy (%)', data=data)
plt.title('Volume of Distribution vs. Drug Efficacy')
plt.xlabel('Volume of Distribution (L)')
plt.ylabel('Drug Efficacy (%)')

plt.subplot(3, 2, 5)
sns.scatterplot(x='Blood Flow Rate (mL/min)', y='Drug Efficacy (%)', data=data)
plt.title('Blood Flow Rate vs. Drug Efficacy')
plt.xlabel('Blood Flow Rate (mL/min)')
plt.ylabel('Drug Efficacy (%)')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

# Set a random seed for reproducibility
np.random.seed(42)

# Number of samples
n = 200

# Generate variables for nanoparticle properties
nanoparticle_size = np.clip(np.random.normal(loc=100, scale=15, size=n), 50, 200)  # Size in nm
nanoparticle_material = np.random.choice(['lipid', 'polymer', 'metal'], size=n)  # Material type
surface_area = np.clip(np.random.uniform(50, 150, n), 0, 200)  # Surface area in m/g
surface_charge = np.clip(np.random.normal(loc=0, scale=10, size=n), -50, 50)  # Charge in mV
release_mechanism_type = np.random.choice(['controlled', 'burst', 'sustained'], size=n)  # Release mechanism
stability = np.random.randint(5, 30, size=n)  # Stability in days

# Generate variables for patient characteristics
patient_age = np.random.randint(18, 90, size=n)  # Age between 18 and 90
patient_gender = np.random.choice(['Male', 'Female'], size=n)  # Gender
patient_weight = np.clip(np.random.normal(loc=70, scale=15, size=n), 40, 150)  # Weight in kg
height = 1.7  # Average height in meters for BMI calculation
patient_bmi = np.clip(patient_weight / (height ** 2), 15, 40)  # BMI in kg/m
pre_existing_conditions = np.random.choice(['Yes', 'No'], size=n)  # Pre-existing conditions
chronic_disease = np.random.choice(['Diabetes', 'Hypertension', 'None'], size=n)  # Chronic disease
liver_function_index = np.clip(np.random.normal(loc=30, scale=10, size=n), 10, 100)  # Liver function index
renal_function_index = np.clip(np.random.normal(loc=90, scale=15, size=n), 30, 150)  # GFR
systolic_bp = np.clip(np.random.randint(90, 180, size=n), 90, 180)  # Systolic BP
diastolic_bp = np.clip(np.random.randint(60, 120, size=n), 60, 120)  # Diastolic BP
heart_rate = np.clip(np.random.normal(loc=70, scale=10, size=n), 50, 100)  # Heart rate in bpm

# Create a DataFrame for all variables
all_data = pd.DataFrame({
    'Nanoparticle Size (nm)': nanoparticle_size,
    'Nanoparticle Material': nanoparticle_material,
    'Surface Area (m/g)': surface_area,
    'Surface Charge (mV)': surface_charge,
    'Release Mechanism Type': release_mechanism_type,
    'Stability (days)': stability,
    'Patient Age (years)': patient_age,
    'Patient Gender': patient_gender,
    'Patient Weight (kg)': patient_weight,
    'Patient BMI (kg/m)': patient_bmi,
    'Pre-Existing Conditions': pre_existing_conditions,
    'Chronic Disease': chronic_disease,
    'Liver Function Index (ALT/AST)': liver_function_index,
    'Renal Function Index (GFR)': renal_function_index,
    'Systolic Blood Pressure (mmHg)': systolic_bp,
    'Diastolic Blood Pressure (mmHg)': diastolic_bp,
    'Heart Rate (bpm)': heart_rate
})

# Save the dataset to a CSV file
all_data.to_csv('all_variables_dataset.csv', index=False)

# Output the first few rows of the generated dataset
print(all_data.head())
```
